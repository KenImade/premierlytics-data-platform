```python
import os
from minio import Minio


def get_minio_client(
    endpoint: str = "localhost:9000",
    access_key: str | None = None,
    secret_key: str | None = None,
    secure: bool = False,
) -> Minio:
    """Return a Minio client connected to the local MinIO instance.

    Credentials fall back to MINIO_ROOT_USER / MINIO_ROOT_PASSWORD env vars
    if not provided explicitly.
    """
    return Minio(
        endpoint=endpoint,
        access_key=access_key or os.environ["MINIO_ROOT_USER"],
        secret_key=secret_key or os.environ["MINIO_ROOT_PASSWORD"],
        secure=secure,
    )


# Usage:
# client = get_minio_client()
# buckets = client.list_buckets()
```
